```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

% matplotlib inline
```

### Radar NIS chart

```python
radar_nis = pd.read_csv('NIS_radar.csv', dtype={'NIS':np.float64})
radar_df = 3
radar_confidence = 0.95
radar_benchmark = stats.chi2.isf(1 - radar_confidence, radar_df)
```

```python
NIS_r = radar_nis.values.ravel()
k_r = radar_nis.index.values
plt.figure(figsize=(12,6))
plt.plot(k_r, NIS_r)
plt.axhline(y=radar_benchmark, color='r', linestyle='--')
plt.xlabel('Timestep')
plt.ylabel('NIS')
plt.show()
```

```python
freq_r = np.sum(NIS_r > radar_benchmark) / len(NIS_r)
print ("Frequency of Radar NIS above Threshold: {:.2f}%".format(freq_r * 100))
```
